```python
import pandas as pd
import numpy as np
```

```python
uncleaned_df = pd.read_csv('pl_transfers_uncleaned_2014_2024_real.csv')
```

```python
uncleaned_df.info()
```

```python
uncleaned_df['Market Value'].unique()
```

```python
# Replace special characters with empty values. Change strings with 'k' or 'm' into float in thousands or millions.

uncleaned_df['Market Value'] = uncleaned_df['Market Value'].str.replace('€', '')
uncleaned_df['Market Value'] = uncleaned_df['Market Value'].str.replace('-', '0')
uncleaned_df.loc[uncleaned_df['Market Value'].str.find('m') >= 1, 'Market Value'] = uncleaned_df[uncleaned_df['Market Value'].str.find('m') >= 1]['Market Value'].str.replace('m', '').astype(float) * 1000000
uncleaned_df.loc[uncleaned_df['Market Value'].str.find('k') >= 1, 'Market Value'] = uncleaned_df[uncleaned_df['Market Value'].str.find('k') >= 1]['Market Value'].str.replace('k', '').astype(float) * 1000



```

```python
uncleaned_df['Market Value'] = uncleaned_df['Market Value'].astype(float)
```

```python
uncleaned_df.info()
```

```python
# Impute unknown Ages (values of '-') with mean.
uncleaned_df['Age'] = uncleaned_df['Age'].str.replace('-', str(uncleaned_df[uncleaned_df['Age'].str.isnumeric()]['Age'].astype(int).mean().astype(int)))
```

```python
uncleaned_df['Age'] = uncleaned_df['Age'].astype(int)
```

```python
uncleaned_df.info()
```

```python
uncleaned_df[~uncleaned_df['Fee'].str.isnumeric()]['Fee'].unique()
```

```python
# Create "is_loan" category to mark records as loans or not loans

uncleaned_df['is_loan'] = np.where(uncleaned_df['Fee'].str.find('loan transfer') >= 0, 1, 0)
```

```python
uncleaned_df[['Fee', 'is_loan']]
```

```python
uncleaned_df['is_loan_return'] = np.where(uncleaned_df['Fee'].str.find('End of loan') >= 0, 1, 0)
```

```python
uncleaned_df[['Fee', 'is_loan_return']]
```

```python
uncleaned_df['Fee'] = uncleaned_df['Fee'].str.replace('€', '')
```

```python
uncleaned_df['Fee'] = uncleaned_df['Fee'].str.replace('free transfer', '0')
```

```python
uncleaned_df['Fee'] = uncleaned_df['Fee'].str.replace('loan transfer', '0')
```

```python
uncleaned_df['Fee'] = np.where(uncleaned_df['Fee'].str.find('End of loan') >= 0, 0, uncleaned_df['Fee'])
```

```python
uncleaned_df['Fee'].unique()
```

```python
uncleaned_df['is_loan'] = np.where(uncleaned_df['Fee'].str.find('Loan fee') >= 0, 1, uncleaned_df['is_loan'])

uncleaned_df.isnull().sum()
```

```python
uncleaned_df['Fee'] = np.where(uncleaned_df['Fee'].str.find('Loan fee') >= 0, uncleaned_df['Fee'].str.replace('Loan fee:\n', ''), uncleaned_df['Fee'])
uncleaned_df.isnull().sum()
```

```python
uncleaned_df['Fee'].unique()
```

```python
# Replace unknown "Fee" values with 0
uncleaned_df['Fee'] = np.where(uncleaned_df['Fee'].str.find('-') >= 0, uncleaned_df['Fee'].str.replace('-', '0'), uncleaned_df['Fee'])
uncleaned_df['Fee'] = np.where(uncleaned_df['Fee'].str.find('?') >= 0, uncleaned_df['Fee'].str.replace('?', '0'), uncleaned_df['Fee'])
```

```python
# Replace strings with numeric values and 'm' or 'k' with floats in millions or thousands
uncleaned_df.loc[uncleaned_df['Fee'].str.find('m') >= 1, 'Fee'] = uncleaned_df[uncleaned_df['Fee'].str.find('m') >= 1]['Fee'].str.replace('m', '').astype(float) * 1000000
uncleaned_df.loc[uncleaned_df['Fee'].str.find('k') >= 1, 'Fee'] = uncleaned_df[uncleaned_df['Fee'].str.find('k') >= 1]['Fee'].str.replace('k', '').astype(float) * 1000
```

```python
uncleaned_df['Fee'] = uncleaned_df['Fee'].astype(float)
```

```python
uncleaned_df.isnull().sum()
```

```python
uncleaned_df.head()
```

```python
uncleaned_df['Season'].unique()
```

```python
#Standardize seasons into YYYY_start/YYYY_end format
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('19/20', '2019/2020')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('20/21', '2020/2021')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('21/22', '2021/2022')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('22/23', '2022/2023')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('23/24', '2023/2024')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('18/19', '2018/2019')
uncleaned_df['Season'] = uncleaned_df['Season'].str.replace('202019/202020', '2019/2020')
uncleaned_df.isnull().sum()
```

```python
# Clean up team values with duplicate "FC"
uncleaned_df['Team'] = uncleaned_df['Team'].str.replace('FC FC', 'FC')
uncleaned_df.to_csv('pl_transfers_2014_2024_cleaned.csv')
```
